# **Import Statements**

```python
import numpy as np
from keras.models import load_model
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from matplotlib import pyplot as plt
```

# **Load the model**

```python
loaded_model = load_model("/content/drive/MyDrive/eff_model.keras")
```

# **Load and preprocess the image**

```python
test_image = load_img("/content/test.jpg", target_size=(224, 224))
test_image_array = img_to_array(test_image)
test_image_array = test_image_array.reshape((1,224,224,3))
```

# **Prediction**

```python
prediction = loaded_model.predict(test_image_array)
predicted_class_index = np.argmax(prediction)
class_names = ['Over Ripen','Under Ripen','Perfect Ripen']
predicted_class_name = class_names[predicted_class_index]
print("predicted class:",predicted_class_name)
```

```python
plt.imshow(image.load_img("/content/test.jpg"))
plt.title(predicted_class_name)
plt.show()
```
